```python
#LIST COMPREHENSIONS TASKS

#1. Task
"""ListComprehension yapısı kullanarak car_crashes verisinde ki numeric değişkenlerin isimlerini büyük harfe çeviriniz ve başına NUM ekleyiniz."""
```

```python
import seaborn as sns
df=sns.load_dataset("car_crashes")
df.columns
```

```python
liste=["NUM_"+i.upper() if df[i].dtype != "O" else i.upper() for i in df.columns]
```

```python
liste
```

```python
#2. Task
"""ListComprehension yapısı kullanarak car_crashes verisinde isminde "no" barındırmayan değişkenlerin isimlerinin sonuna "FLAG" yazınız."""
```

```python
liste2=[i.upper() if "no" in i else i.upper()+"_FLAG" for i in df.columns]
```

```python
liste2
```

```python
#3. Task
"""ListComprehension yapısı kullanarak aşağıda verilen değişken isimlerinden FARKLI olan değişkenlerin isimlerini seçiniz ve yeni bir data frame oluşturunuz."""
```

```python
ogg_list=["abbrev","no_previous"]
```

```python
new_cols=[]
```

```python
[i if i in ogg_list else new_cols.append(i) for i in df.columns]
```

```python
new_cols
```

```python
new_df = df[new_cols]
```
